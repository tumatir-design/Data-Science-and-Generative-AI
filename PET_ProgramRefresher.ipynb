```python
import os
import csv
import datetime
from typing import List, Dict, Union, Optional

# Personal Expense Tracker
class ExpenseTracker:
    def __init__(self):
        self.expenses: List[Dict[str, Union[str, float]]] = []
        self.budget: Optional[float] = None
        self.data_file = "expenses.csv"
        self.load_expenses()
        
    #1. Add an expense:    
    def add_expense(self) -> None:
        """Prompt user for expense details and add to expenses list"""
        print("\n=== Add New Expense ===")
        
        # Get date with validation
        while True:
            date_str = input("Enter date (YYYY-MM-DD): ")
            try:
                # Validate date format
                datetime.datetime.strptime(date_str, "%Y-%m-%d")
                break
            except ValueError:
                print("Invalid date format. Please use YYYY-MM-DD.")
    
        # Get category
        category = input("Enter category (e.g., Food, Travel): ")
       
        # Get amount with validation
        while True:
            amount_str = input("Enter amount spent: ")
            try:
                amount = float(amount_str)
                if amount <= 0:
                    print("Amount must be greater than zero.")
                    continue
                break
            except ValueError:
                print("Invalid amount. Please enter a number.")
        
        description = input("Enter a brief description: ")
        
        # Create and add expense
        expense = {
            'date': date_str,
            'category': category,
            'amount': amount,
            'description': description
        }
        
        self.expenses.append(expense)
        print("Expense added successfully!")  

        # Auto-save after adding an expense
        self.save_expenses(quiet=True)  
        
    # 2. View expenses:     
    def view_expenses(self) -> None:
        """Display all stored expenses"""
        if not self.expenses:
            print("\nNo expenses recorded yet.")
            return
        
        print("\n=== Expense Records ===")
        print(f"{'Date':<10} {'Category':<10} {'Amount':<10} {'Description':<50}")
        print("-" * 67)
        
        for expense in self.expenses:
            # Validate expense has all required fields
            if all(key in expense for key in ['date', 'category', 'amount', 'description']):
                print(f"{expense['date']:<12} {expense['category']:<15} ${expense['amount']:<9.2f} {expense['description']:<30}")
            else:
                print(f"Skipping incomplete expense record: {expense}")
        
        # Show total
        total = sum(expense['amount'] for expense in self.expenses if 'amount' in expense)
        print("-" * 67)
        print(f"Total expenses: ${total:.2f}")

    #3. Set and track the budget:
    def set_budget(self) -> None:
        """Set monthly budget"""
        print("\n=== Set Monthly Budget ===")
        while True:
            budget_str = input("Enter your monthly budget amount: ")
            try:
                budget = float(budget_str)
                if budget <= 0:
                    print("Budget must be greater than zero.")
                    continue
                self.budget = budget
                print(f"Monthly budget set to ${self.budget:.2f}")
                break
            except ValueError:
                print("Invalid amount. Please enter a number.")
    
    def track_budget(self) -> None:
        """Track expenses against budget"""
        if self.budget is None:
            print("\nNo budget set. Please set a budget first.")
            self.set_budget()
            
        total_expenses = sum(expense['amount'] for expense in self.expenses if 'amount' in expense)
        remaining = self.budget - total_expenses
        
        print("\n=== Budget Tracking ===")
        print(f"Monthly Budget: ${self.budget:.2f}")
        print(f"Total Expenses: ${total_expenses:.2f}")
        print(f"Remaining: ${remaining:.2f}")
        
        if remaining < 0:
            print("WARNING: You have exceeded your budget!")
            
    # 4  to implement the save_expenses method
    def save_expenses(self, quiet=False) -> None:
        """Save expenses to CSV file"""
        try:
            with open(self.data_file, 'w', newline='') as file:
                if not self.expenses:
                    if not quiet:
                        print("No expenses to save.")
                    return
                    
                # Get all possible field names from all expenses
                fieldnames = set()
                for expense in self.expenses:
                    fieldnames.update(expense.keys())
                
                writer = csv.DictWriter(file, fieldnames=list(fieldnames))
                writer.writeheader()
                writer.writerows(self.expenses)
                
                if not quiet:
                    print(f"Expenses saved to {self.data_file}")
        except Exception as e:
            print(f"Error saving expenses: {e}")
    
    def load_expenses(self) -> None:
        """Load expenses from CSV file"""
        if not os.path.exists(self.data_file):
            print(f"No existing expense data found.")
            return
            
        try:
            with open(self.data_file, 'r', newline='') as file:
                reader = csv.DictReader(file)
                self.expenses = []
                for row in reader:
                    # Convert amount from string to float
                    if 'amount' in row:
                        row['amount'] = float(row['amount'])
                    self.expenses.append(row)
            print(f"Loaded {len(self.expenses)} expenses from {self.data_file}")
        except Exception as e:
            print(f"Error loading expenses: {e}")
            
    #5. Create an interactive menu:    
    def display_menu(self) -> None:
        """Display the main menu"""
        print("\n=== Expense Tracker Menu ===")
        print("1. Add expense")
        print("2. View expenses")
        print("3. Track budget")
        print("4. Save expenses")
        print("5. Exit")
    
    def run(self) -> None:
        """Main program loop"""
        print("Welcome to Expense Tracker!")
        
        while True:
            self.display_menu()
            choice = input("\nEnter your choice (1-5): ")
            
            if choice == '1':
                self.add_expense()
            elif choice == '2':
                self.view_expenses()
            elif choice == '3':
                self.track_budget()
            elif choice == '4':
                self.save_expenses()
            elif choice == '5':
                self.save_expenses()
                print("Thank you for using Expense Tracker. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 5.")


# Run the expense tracker
if __name__ == "__main__":
    tracker = ExpenseTracker()
    tracker.run()
```
